# Polaris Catalog Demo

This notebook demonstrates how to connect to the **Polaris Catalog** (an Iceberg REST Catalog) using Spark.

**Prerequisites:**
- Polaris service running in K8s
- MinIO service running in K8s
- Spark Operator and RBAC configured

```python
import connector

# Initialize Spark Session using our helper module
spark = connector.create_spark_session("polaris-demo")
print("Spark Session Created!")
```

## 1. Manage Namespaces

In Iceberg/Polaris, a **Namespace** is equivalent to a Database/Schema in traditional SQL.

```python
# Create a new namespace 'demo_ns'
print("Creating namespace 'polaris.demo_ns'...")
spark.sql("CREATE DATABASE IF NOT EXISTS polaris.demo_ns").show()
```

```python
# List all namespaces in the Polaris catalog
print("Listing namespaces in Polaris:")
spark.sql("SHOW NAMESPACES IN polaris").show()
```

```python
# Describe the namespace to see properties (like location)
spark.sql("DESCRIBE DATABASE EXTENDED polaris.demo_ns").show(truncate=False)
```
